```python
# Uppercase Converter
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
uppercased = [fruit.upper() for fruit in fruits]
print(uppercased)   # ['APPLE', 'BANANA', 'CHERRY', 'DATE', 'ELDERBERRY']
print(fruits) 
```

```python
# Flatten a 2D List
matrix = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flat = [item for row in matrix for item in row]
print(flat)
```

# FizzBuzz with Comprehension
Produce the classic FizzBuzz sequence for integers 1 to 50 
in a single list comprehension. Each element should be 
"FizzBuzz" if the number is divisible by both 3 and 5, "Fizz" if divisible by 3 only, "Buzz" if divisible by 5 only, and the number itself (as an integer) otherwise.


```python
fizzbuzz = [
    "FizzBuzz" if n % 15 == 0 else
    "Fizz"     if n % 3  == 0 else
    "Buzz"     if n % 5  == 0 else
    n
    for n in range(1, 51)
]

print(fizzbuzz)
```

```python
#  Cartesian Product
xs = [1, 2, 3]
ys = ["a", "b", "c"]
product = [(x, y) for x in xs for y in ys]
for pair in product:
    print(pair, end="  ")
```

```python
# Extract Digits
strings = ["abc123", "hello", "42px", "year2024", "no-digits"]

digits = [char for s in strings for char in s if char.isdigit()]
```

```python
# Nested Filtering

groups = [[1, 2, 3], [-1, 4, 5], [6, 7, 8], [0, 9, 10], [-3, -1, 2], [4, 5, 6]]

all_positive = [group for group in groups if all(n > 0 for n in group)]

print(all_positive)
```

```python
# Conditional Dict Comprehension
scores = {
    "Alice":   82,
    "Bob":     45,
    "Charlie": 91,
    "Diana":   37,
    "Eve":     55,
    "Frank":   49,
}

passed = {name: score for name, score in scores.items() if score >= 50}

# print(passed)
```

```python
# Set of Common Elements

list_a = [1, 2, 3, 4, 5, 3, 2]
list_b = [3, 4, 5, 6, 7, 4, 5]

common = {n for n in list_a if n in list_b}

print(common)
```

```python
# Character Frequency Set
sentence = "comprehension makes python powerful"

repeated = {char for char in sentence if char != " " and sentence.count(char) > 1}

print(sorted(repeated))
```
