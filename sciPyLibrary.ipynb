```python
# Basic Functions
from scipy import integrate
import numpy as np
# Integration is a technique used for calculating the 
# integral of a function when an analytical solution
# is difficult or impossible to obtain.

f = lambda x: x**2
result, error = integrate.quad(f, 0, 1)
print(result)
```
