```python
# unordered # Key:value
dict1 = dict()
dict2 = {}
dict3 = {1:'a',2:'b',3:'c'}
print(dict1, dict2, dict3)
dict4 = {"z":34, "t": 98, "A":78, "B":67,"C":87}
print(dict4)
print(dict4["z"])
###### key 
for i in dict4:
    print(i, end=" ")  # key
print() 
# #####  dict[Key] 
for i in dict4:
    print(i, dict4[i],sep=":",end=" , ")
```

```python
print(dict4.items())

for i, j in dict4.items():
    print(i, j,sep=":")
    
# t1 = (1,2)
# a , b = t1 
# print(a, b)
```
